# **Extracting Tesla Stock Data Using yfinance**

```python
!pip install yfinance
import yfinance as yf

# Define the ticker symbol for Tesla
ticker_symbol = "TSLA"

# Get the stock data for Tesla
tesla_data = yf.Ticker(ticker_symbol)

# Get the historical stock prices for the past year
hist = tesla_data.history(period="1y")

# Print the historical stock prices
print(hist)
```

# **Extracting Tesla Revenue Data Using Webscraping**

```python
# Install libraries if needed
!pip install beautifulsoup4
!pip install requests

# Import necessary libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the Wikipedia page containing Tesla revenue data
url = "https://en.wikipedia.org/wiki/Tesla,_Inc."  # Replace with the actual URL

# Send an HTTP GET request to the URL
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing revenue data (inspect the Wikipedia page to identify the table)
revenue_table = soup.find('table', class_='wikitable')  # Replace with the actual table class

# Extract data from the table
data = []
for row in revenue_table.find_all('tr')[1:]:  # Skip the header row
    columns = row.find_all('td')
    if len(columns) >= 2:
        year = columns[0].text.strip()
        revenue = columns[1].text.strip()
        data.append([year, revenue])

# Create a Pandas DataFrame from the extracted data
df = pd.DataFrame(data, columns=['Year', 'Revenue'])

# Display the DataFrame
print(df)
```

```python
df.tail(5)
```
